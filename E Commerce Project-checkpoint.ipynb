```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Sample - Superstore.csv", encoding= "latin-1")
```

# Converting Date Columns

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
df["Order Month"] = df["Order Date"].dt.month
df["Order Year"] = df["Order Date"].dt.year
df["Order Day of Week"] = df["Order Date"].dt.dayofweek
```

```python
df.info()
```

```python
df.head(5)
```

# Monthly sales Analysis

```python
monthly_Sales = df.groupby("Order Month")["Sales"].sum().reset_index()
```

```python
sns.lineplot(data= monthly_Sales, x= "Order Month", y= "Sales", marker= "o")
plt.title("Monthly Sales Analysis")
plt.grid()
plt.show()
```

# Sales By Category

```python
sales_by_category = df.groupby("Category")["Sales"].sum().reset_index()

plt.pie(
    sales_by_category["Sales"],
    labels=sales_by_category["Category"],
    colors=["gold", "coral", "green"],
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Sales By Category")
plt.axis('equal')
plt.show()
```

# Sales Analysis by Subcategory

```python
sales_by_subcategory = df.groupby("Sub-Category")["Sales"].sum().reset_index()

plt.figure(figsize=(22, 8))
ax = sns.barplot(data= sales_by_subcategory, x= "Sub-Category", y= "Sales")

for bar in ax.containers:
    ax.bar_label(bar)

plt.title("Sales By Sub-Category")
plt.tight_layout()
plt.show()
```

# Monthly Profit Analysis

```python
profit_by_month = df.groupby('Order Month')["Profit"].sum().reset_index()

plt.figure(figsize=(16, 6))
ax = sns.barplot(data= profit_by_month, y= 'Order Month', x= "Profit")

for bar in ax.containers:
    ax.bar_label(bar)

plt.title("Monthly Profit Analysis")
plt.tight_layout()
plt.show()
```

# Profit By Category

```python
profit_by_category = df.groupby("Category")["Profit"].sum().reset_index()

plt.pie(
    profit_by_category["Profit"],
    labels=profit_by_category["Category"],
    colors=["red", "green", "orange"],
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Profit By Category")
plt.tight_layout()
plt.show()
```

# Profit By Sub-Category

```python
profit_by_subcategory = df.groupby("Sub-Category")["Profit"].sum().reset_index()

plt.figure(figsize= (16,8))
ax = sns.barplot(data= profit_by_subcategory, x= "Sub-Category", y= "Profit")

for bar in ax.containers:
    ax.bar_label(bar)

plt.title("Profit By Sub-Category")
plt.tight_layout()
plt.show()
```

```python
df.columns
```

# Sales and Profit - Cutomer Segment

```python
sales_profit_segment = df.groupby("Segment")[["Sales", "Profit"]].sum().reset_index()

sns.barplot(data= sales_profit_segment, x= "Sales", y= "Profit", hue= "Segment")
plt.title("Customer Segment (Sales and Profit)")
plt.tight_layout()
plt.show()
```

# Sales to Profit Ratio

```python
sales_profit_segment = df.groupby("Segment")[["Sales", "Profit"]].sum().reset_index()

sales_profit_segment["Sales_to_Profit_Ratio"] = sales_profit_segment["Sales"] / sales_profit_segment["Profit"]
print(sales_profit_segment[["Segment", "Sales_to_Profit_Ratio"]])
```
